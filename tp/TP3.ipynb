```python
## Fct qui qui récupérer les questions et leurs mots clés pour les mettre dans une list
def parcfct(data,lang):
    quest = []
    keywords = []
    for questions in data:
        for question in questions:
            if question.attrib.get('lang') == lang :
                if question.tag == "string" :
                    #print("QCT : " + str(question.text))
                    quest.append(question.text)
                elif question.tag == "keywords" :
                    #print("KEYWORDS : " + str(question.text))
                    keywords.append(question.text)
    return quest,keywords
```

```python
## Fct qui tokenise la liste de questions et attribut des tag aux mots
def quest_tag(data):
    tag_quest = []
    for i in range(len(data)):
        tag_quest.append(nltk.pos_tag(nltk.word_tokenize(data[i])))
    return tag_quest
```

```python
##Fct qui trouve les mots clés
def found_keyswords(data):
    keywords = []
    others = []
    for i in range(len(data)):
        liste = []
        liste2 = []
        for j in range(len(data[i])):   
                
            if data[i][j][1] == 'NNP' and data[i][j][0] != 'Which' :
                liste.append(data[i][j][0])
                
            elif data[i][j][1] == 'NN':
                if j > 0 and data[i][j-1][1] == 'JJS':
                    liste.append(data[i][j-1][0])
                liste.append(data[i][j][0])
            elif data[i][j][1] == 'NNS':
                liste.append(data[i][j][0])
                
            elif data[i][j][1] == 'NNPS':
                liste.append(data[i][j][0])
                
            elif data[i][j][1] == 'VBD' or data[i][j][1] == 'VBZ' or data[i][j][1] == 'DT':
                if data[i][j][0] != "is" and data[i][j][0] != "was" and data[i][j][0] != "are" and data[i][j][0] != "did" and data[i][j][0] != "does" and data[i][j][0] != "the" :
                    liste.append(data[i][j][0])
            else :
                liste2.append(data[i][j][0])
        keywords.append(liste)
        others.append(liste2)
    return keywords, others
```

```python
## Fct qui trouve les mots clés (spécialisée pour les question commencant avec which)
def found_which(list_tag):
    for i in range(len(list_tag)):
        if list_tag[i][0] == "which" or list_tag[i][0] == "Which" :
            dbo = list_tag[i+1][0]
    return dbo
```

```python
## Ici on cherche si des mots constitue par exemple une organisation ou une personne 
def found_chunk(list_tag):
    named_entities = ""
    parse_tree = nltk.chunk.ne_chunk(list_tag)
    #for l in range(len(label_possible)):
    for t in parse_tree.subtrees():
        if t.label() != 'S':
            t_list= list(t)
            for i in range(len(t_list)):
                if t_list[i][0].find('Which') == -1 :
                    if len(named_entities)> 0 :
                        named_entities = named_entities + "_" + t_list[i][0]
                    else :
                        named_entities = t_list[i][0]
    return named_entities
```

```python
# Cette fct retire les mots clés qu'on considére deja comme des ressources (organisation/personne...)
def ressources(keylist,list_tag):
    chunk = found_chunk(list_tag)
    ressources = ""
    for i in range(len(keylist)):
        if chunk.find(str(keylist[i])) == -1 :
            ressources = ressources + " " + keylist[i]            
    return chunk, ressources
```

```python
##PAS EFFICACE DU TOUT
## FCt qui regarde si les mots clés trouvés correspondent aux mots dans le dictionnaire "relation.txt"
def ismatching(relation,keyword):
    key = keyword.split(" ")
    sc = 0
    sckey = [0]*len(key)
    db = [""]*len(key)
    for i in range(len(key)):
        for j in range(len(relation[1])):
            synsets1 = wordnet.synsets(key[i])
            synsets2 = wordnet.synsets(relation[1][j])
            if len(synsets1) != 0 and len(synsets2) != 0 :
                sc = synsets1[0].path_similarity(synsets2[0])
            else :
                sc = 0
            if sc is None :
                sc = 0
            if sckey[i] < sc :
                db[i] = relation[0][j]+":"+relation[1][j]
                sckey[i] = sc
    id = sckey.index(max(sckey))
    return db[id]
```

```python
## Nouvelle version de matching mais en plus efficace
def ismatching2(relation,keyword):
    key = keyword.split(" ")
    sc = 0
    sckey = [0]*len(key)
    db = [""]*len(key)
    for i in range(len(key)):
        for j in range(len(relation[1])):
            sc = difflib.SequenceMatcher(None,key[i],relation[1][j]).ratio()
            if sc is None :
                sc = 0
            if sc > sckey[i] :
                db[i] = relation[0][j]+":"+relation[1][j]
                sckey[i] = sc
        if sckey[i] <= 0.6 :
            for j in range(len(relation[1])):
                synsets1 = wordnet.synsets(key[i])
                synsets2 = wordnet.synsets(relation[1][j])
                if len(synsets1) != 0 and len(synsets2) != 0 :
                    sc = synsets1[0].path_similarity(synsets2[0])
                else :
                    sc = 0
                if sc is None :
                    sc = 0
                if sckey[i] < sc :
                    db[i] = relation[0][j]+":"+relation[1][j]
                    sckey[i] = sc
    id = sckey.index(max(sckey))
    return db[id]
```

```python
##La fct qui va extraire les relations du doc "relation.txt"
def extract_txt(txt):
    doc = txt + ".txt"
    dptype, word = [], []
    with open(doc, "r") as f_read:
        for line in f_read:
            line = line.strip()
            if line:
                ti, vi = [str(mot) for mot in line.split(":")]
                dptype.append(ti), word.append(vi)
    return dptype, word
relation = extract_txt("relations")
```

```python
##Fct qui ecrit la query
def ecriture_query(ress,dbo, typeq):
    #string
    #date
    #number
    if len(typeq) > 0 :
        uri = typeq
    else :
        uri = "uri" 
    query = "PREFIX dbo: <http://dbpedia.org/ontology/> PREFIX res: <http://dbpedia.org/resource/> SELECT DISTINCT ?"+uri+" WHERE { res:" + ress + " "+dbo+ " ?"+uri+" . }"
    return query
```

```python
#Fct qui permet également de trier les ressources
def who_ress(list_tag,ressdp):
    dbo = ""
    for i in range(len(list_tag)):
        if list_tag[i][1] == 'VBD' or list_tag[i][1] == 'VBZ' or list_tag[i][1] == 'VBG':
            dbo = list_tag[i][0]
        ressdp.lstrip(dbo)
    ressdb = ressdp.replace(" ", "_")
    return dbo,ressdb
```

```python
## FCt qui envoie la requete query et qui print les resultats
def query_requ(query,f):
    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query()
    ##J'ai pas ressi a récuperer les réponses pour faire l'évaluation
    ##f.write(str(results) + "\n")
    results.print_results()
    print()
```
